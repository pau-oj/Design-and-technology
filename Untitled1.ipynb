```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import scipy.signal as signal # Importamos para suavizar la seÃ±al

def generar_forma_suave(ruta_audio, archivo_salida='forma_suave.svg'):
    print(f"Procesando el audio para forma suave: {ruta_audio}...")

    # 1. Cargar audio (primeros 5 segundos)
    y, sr = librosa.load(ruta_audio, sr=None, duration=5.0)

    # 2. Extraer SOLAMENTE la Amplitud (Volumen)
    # Usamos un tamaÃ±o de ventana (frame_length) mÃ¡s grande para que la curva de volumen sea mÃ¡s suave desde el principio
    rms = librosa.feature.rms(y=y, frame_length=2048, hop_length=512)[0]

    # Opcional: Aplicar un filtro extra para suavizar aÃºn mÃ¡s la curva de volumen y evitar cambios bruscos
    # Esto ayuda a que la forma final sea muy "lÃ­quida"
    rms_suavizado = signal.savgol_filter(rms, window_length=51, polyorder=3)
    # Asegurar que no haya valores negativos despuÃ©s del filtro
    rms_suavizado = np.clip(rms_suavizado, 0, None)

    # 3. Normalizar (valores de 0 a 1)
    rms_norm = (rms_suavizado - np.min(rms_suavizado)) / (np.max(rms_suavizado) - np.min(rms_suavizado) + 1e-6)

    # 4. GeometrÃ­a Circular
    n_points = len(rms_norm)
    theta = np.linspace(0, 2 * np.pi, n_points, endpoint=False)

    # ParÃ¡metros fÃ­sicos (en mm)
    radio_base = 50.0 # El tamaÃ±o mÃ­nimo del cÃ­rculo
    altura_maxima = 40.0 # CuÃ¡nto se expande el cÃ­rculo en la parte mÃ¡s ruidosa

    # El radio final es solo la base mÃ¡s la expansiÃ³n por volumen
    r = radio_base + (rms_norm * altura_maxima)

    # Aplicar ventana para asegurar que el cÃ­rculo cierre perfectamente sin un salto
    ventana = np.hanning(n_points)
    # Esta mezcla hace que el inicio y el fin se acerquen suavemente al radio base
    r = radio_base + ((r - radio_base) * ventana)

    # Convertir a coordenadas cartesianas para dibujar
    x = r * np.cos(theta)
    y_cart = r * np.sin(theta)

    # Cerrar el polÃ­gono
    x = np.append(x, x[0])
    y_cart = np.append(y_cart, y_cart[0])

    # 5. Graficar y Exportar
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.plot(x, y_cart, color='#00695c', linewidth=3) # LÃ­nea mÃ¡s gruesa y suave
    ax.fill(x, y_cart, color='#4db6ac', alpha=0.4)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.title("TraducciÃ³n de Volumen a Forma Suave\n(Sin texturas de frecuencia)")

    plt.savefig(archivo_salida, format='svg', bbox_inches='tight', transparent=True)
    print(f"Â¡Forma suave generada! Descarga el archivo: {archivo_salida}")
    plt.show()

# === CÃMO USARLO EN COLAB ===
# 1. Sube tu archivo de audio al panel de la izquierda.
# 2. Cambia 'tu_audio.wav' por el nombre real de tu archivo.
generar_forma_suave('ARROZ.mp4', 'mi_forma_circular.svg')
```
